```python
import numpy as np
import matplotlib.pyplot as plt
```

Function that uses inverse transform sampling to generate a sequence of 1000 numbers

```python
# Function to generate random variables
def inverse_transform_sampling(n):
    random_numbers = []
    for _ in range(n):
        u = np.random.uniform(0, 1)  # Generate a uniform random number between 0 and 1
        if u <= 0.25:
            # If u is in the interval [2, 3] use inverted cdf
            x = 2 + 2 * np.sqrt(u)
        else:
            # If u is in the interval (3, 6] use inverted cdf
            x = 6 - 3 * np.sqrt(2 * (1 - u))
        random_numbers.append(x)
    return np.array(random_numbers)

# Generating a sequence of 1000 numbers by calling the function
n_samples = 1000
samples = inverse_transform_sampling(n_samples)
```

histogram generation

```python
plt.hist(samples, bins=30, density=True, edgecolor='black', alpha=0.7)
plt.title('Histogram Random Variation Generator')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
